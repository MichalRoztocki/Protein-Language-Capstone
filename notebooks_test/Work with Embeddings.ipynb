### Creating files with embeddings

```python
file_name = '../data_test/data_h5/train_small1_dlm_avg.h5'
```

```python
proteins = {}

with h5py.File(file_name, 'r') as hf:
    for key in hf.keys():
        print(key, end=' ')
        data = hf.get(key)
        n = np.array(data)
        proteins[key] = n
```

```python
proteins
```

```python
df = pd.DataFrame(proteins).T
```

```python
df.index.name = 'Sequence_id'
```

```python
df
```

Save the file for unsupervised learning tasks

```python
df.to_csv('seq_emb.tsv', sep='\t' )
```

Save the files for Embedding Projector

```python
df.index.to_frame(index=False).to_csv('Sequence_ids.tsv', index=False, sep='\t' )
```

```python
df.to_csv('Embeddings.tsv', index=False, sep='\t' )
```

The files Sequence_ids.tsv and 'Embedings.tsv are for now located in the notebooks_test.

### DNA Binding Proteins

```python
file_name = '../data/dna_binding/train_h5/train_dlm_sum.h5'
```

```python
proteins = {}

with h5py.File(file_name, 'r') as hf:
    for key in hf.keys():
        data = hf.get(key)
        n = np.array(data)
        proteins[key] = n
```

```python
df_dna = pd.DataFrame(proteins).T
# parquet must have string column names
df_dna.columns = df_dna.columns.map(str)
```

```python
df_dna.index.name = 'Sequence_id'
```

```python
df_dna
```

Extract the label column from the original csv file

```python
csv_name = '../data/dna_binding/train.csv'
```

```python
dna_cl = pd.read_csv(csv_name)
dna_cl.columns
```

```python
dna_cl.info()
```

After reading data from h5 file, data is sorted by sequence id (h5 key).   
For supervised learning tasks, we will sort csv data by sequence id as well, and keep only the label column.

```python
dna_cl = dna_cl.sort_values(by=['code']).reset_index(drop=True)['label']
```

Concatenate (merge) the label column to sequence ids and embeddings

```python
df_dna = df_dna.reset_index().assign(label=dna_cl)
```

```python
df_dna
```

Write dataframe to csv file

```python
df_dna.to_csv('../data/dna_binding/train_parquet/dna_train_dlm_sum_all.csv', index=False)
```
